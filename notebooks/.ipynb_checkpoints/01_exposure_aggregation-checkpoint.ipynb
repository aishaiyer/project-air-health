```python
import os
import numpy as np
import pandas as pd
import seaborn
# --- Sanity-check plot: seasonality across counties ---
import matplotlib.pyplot as plt
import seaborn as sns


PROJECT_ROOT = os.path.abspath(os.path.join(os.getcwd(), ".."))
DATA_RAW = os.path.join(PROJECT_ROOT, "data", "raw")
DATA_PROCESSED = os.path.join(PROJECT_ROOT, "data", "processed")

PM25_DAILY_PATH = os.path.join(DATA_RAW, "pm25_daily_dc_md.csv")
df_pm = pd.read_csv(PM25_DAILY_PATH, low_memory=False)

print("Loaded:", df_pm.shape)
print([c for c in df_pm.columns if "state" in c.lower() or "county" in c.lower()][:50])
df_pm.head()
```

```python
unique_counties = (
    df_pm[["State Name", "County Name"]]
    .drop_duplicates()
    .sort_values(["State Name", "County Name"])
    .reset_index(drop=True)
)

print("Unique (State, County) pairs:", len(unique_counties))
unique_counties
```

```python
# --- DMV-only GEOID mapping + county-month aggregation ---

# 1) Hardcoded DMV county GEOID mapping (documented + explicit)
county_to_geoid = {
    ("District of Columbia", "District of Columbia"): "11001",

    ("Maryland", "Baltimore"): "24005",
    ("Maryland", "Baltimore (City)"): "24510",
    ("Maryland", "Cecil"): "24015",
    ("Maryland", "Dorchester"): "24019",
    ("Maryland", "Garrett"): "24023",
    ("Maryland", "Harford"): "24025",
    ("Maryland", "Howard"): "24027",
    ("Maryland", "Kent"): "24029",
    ("Maryland", "Montgomery"): "24031",
    ("Maryland", "Prince George's"): "24033",
    ("Maryland", "Washington"): "24043",
}

# 2) Assign GEOID
df_pm["geoid"] = df_pm.apply(
    lambda r: county_to_geoid.get((r["State Name"], r["County Name"])),
    axis=1
)

# Sanity check
missing = df_pm[df_pm["geoid"].isna()][["State Name", "County Name"]].drop_duplicates()
print("Missing GEOID mappings:")
print(missing if not missing.empty else "None")

# 3) Parse date + month
df_pm["date"] = pd.to_datetime(df_pm["date"])
df_pm["month"] = df_pm["date"].dt.to_period("M").dt.to_timestamp()

# 4) Aggregate to county Ã— month
y_col = "Arithmetic Mean"

county_month = (
    df_pm.groupby(["geoid", "month"])
         .agg(
             pm25_median=(y_col, "median"),
             pm25_mean=(y_col, "mean"),
             n_obs=(y_col, "size"),
         )
         .reset_index()
)

print("County-month table:", county_month.shape)
county_month.head()
```

```python
# --- Save county-month exposure table ---
OUT_PATH = os.path.join(DATA_PROCESSED, "pm25_county_monthly_dmv.csv")
county_month.to_csv(OUT_PATH, index=False)
print("Saved:", OUT_PATH)

plt.figure(figsize=(9, 4))
sns.lineplot(
    data=county_month,
    x="month",
    y="pm25_median",
    hue="geoid",
    legend=False,
    alpha=0.7
)

plt.title("County-Level PM2.5 Seasonality (Median, DMV)")
plt.xlabel("Month")
plt.ylabel("PM2.5 (Âµg/mÂ³)")
plt.tight_layout()
plt.show()
```
